```python
import pandas as pd
import duckdb
import os

# Create database folder
os.makedirs(
    "../database",
    exist_ok=True
)

# Connect to DB
conn = duckdb.connect(
    "../database/airport_ops.duckdb"
)

# Read airport dataset
url = "https://ourairports.com/data/airports.csv"

airports = pd.read_csv(url)

# Keep medium + large airports
airports = airports[
    airports["type"].isin(
        ["large_airport", "medium_airport"]
    )
]

# Register dataframe
conn.register(
    "airports_df",
    airports
)

# Create table
conn.execute("""
CREATE OR REPLACE TABLE airports AS
SELECT *
FROM airports_df
""")

print("Airport table created!")

print(
    conn.sql("SHOW TABLES")
    .df()
)
```

```python
conn.close()
```
